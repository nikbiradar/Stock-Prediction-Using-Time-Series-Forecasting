```python
import os
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
df=pd.read_csv('TSLA.csv')
df_close=df['Close']
df_close.plot()
len(df)
```

```python
def stationarity(df):
    rollmean=df.rolling(12).mean()
    rollstd=df.rolling(12).std()
    df.plot(color='black')
    rollmean.plot(color='blue')
    rollstd.plot(color='red')
    adf=adfuller(df,autolag='AIC')
    output=pd.Series(adf[:4],index=['test stats','p-value','no of lags','no of observations'])
    for key,val in adf[4].items():
        output['Critical value (%s)' %key]=val
    print(output)
    
stationarity(df_close)
```

```python
train,test=df_close[0:int(len(df)*0.7)], df_close[int(len(df)*0.7):]
# m=auto_arima(train,start_p=0,start_q=0,test='adf',m=1,d=None,D=0,start_P=0,
#                 seasonal=False,trace=True,
#                 error_action='ignore',
#                 suppress_warnings=True,
#                 stepwise=True)
# m.summary()
```

```python
his=[x for x in train]
pred=[]
n=len(test)
for entry in test:
    model=ARIMA(his,order=(1,1,0))
    fitted=model.fit(disp=0)
    out=fitted.forecast()
    yhat=out[0]
    pred.append(yhat[0])
    his.append(entry)
```

```python
test_set_range = df[int(len(df)*0.7):].index
plt.plot(test_set_range, pred, color='blue', marker='o', linestyle='dashed',label='Predicted Price')
plt.plot(test_set_range, test, color='red', label='Actual Price')
plt.title('TESLA Prices Prediction')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.xticks(np.arange(2118,3025,50), df.Date[2118:3025:50])
plt.legend()
plt.show()
```
