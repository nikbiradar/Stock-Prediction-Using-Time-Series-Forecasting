```python
from nsepy.derivatives import get_expiry_date
```

```python
stock=get_history(symbol='SBIN',start=date(2015,1,1),end=date(2015,1,10),futures=True,expiry_date=list(get_expiry_date(2015,1))[0])
stock['Close'].plot(color='red')
stock['Open'].plot(color='black')
```
